```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
tally_data = {"Heritage":9, "Jersey": 5, "Amul": 14, "Vijaya":7}
keys_list = list(tally_data.keys())
values_list = list(tally_data.values())
```

```python
fig = plt.figure(figsize = (10, 4))
plt.bar(keys_list, values_list, color ='orange',
        width = 0.5)
plt.xlabel("Diary products")
plt.ylabel("Usage of dairy brands/person")
plt.title("Dairy Product usage")
plt.show()
```
